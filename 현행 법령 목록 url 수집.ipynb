```python
import pandas as pd
import xml.etree.ElementTree as ET # xml 문서를 파싱(parsing)하고 검색할 때 사용
from urllib.request import urlopen
from tqdm import trange #  진행사항 확인을 위한  Progress Bar
```

```python
url = 'https://www.law.go.kr/DRF/lawSearch.do?OC=skyggg3&target=law&type=XML' # url 선언
response = urlopen(url).read() # 데이터 받아오기
response
```

```python
xtree = ET.fromstring(response) # 문자열 값으로 존재하는 경우 fromstring() 메서드에 xml 데이터 문자열을 넘긴다
```

```python
totalCnt = int(xtree.find('totalCnt').text) # 반복횟수를 정해주기 위해 전체 갯수 확인
totalCnt
```

```python
items = xtree[8:] # 파싱할 요소에 접근
items
```

```python
 # 상세링크 text를 받아오는지 확인
for node in items:
  법령상세링크 = node.find('법령상세링크').text # 상세링크 text를 받아오는지 확인
  print(법령상세링크)
```

```python
# 최종 코드
url = 'https://www.law.go.kr/DRF/lawSearch.do?OC=skyggg3&target=law&type=XML'
response = urlopen(url).read()
xtree = ET.fromstring(response)

totalCnt = int(xtree.find('totalCnt').text)

num = 0
page = 1
rows = []

for i in trange(int(totalCnt / 20)):
    try:
        items = xtree[8:]
    except:
        break
        
    for node in items:
        num += 1
        법령상세링크 = node.find('법령상세링크').text
        print("순번:",num, 법령상세링크)

        rows.append({'법령상세링크': 법령상세링크})
   
    page += 1
    print(page, "page")
    url = "https://www.law.go.kr/DRF/lawSearch.do?OC=skyggg3&target=law&page={}&type=XML".format(page) # 페이지를 바꿔서 파싱
    response = urlopen(url).read()
    xtree = ET.fromstring(response)
urls = pd.DataFrame(rows)
urls.to_csv('urls.csv', index=False)
```

총 수집된 url은 5160개로 맨 마지막 페이지의 9개 url이 누락되었음을 확인. 수기로 추가해주고 엑셀에서 중복된 항목 제거 기능을 사용하여 중복값이 없는지 확인까지 완료함.
